## **Task 1 (20 marks): Why do employee leave?**

* Expensive in terms of both money and time to train new employees
* Loss of experienced employees
* Impact on productivity
* Impact on profit

1.2 Is there a difference in attrition between male and femal employees?

```python
groupby_attrition_gender = df_attrition.groupby(["Attrition", "Gender"]).agg(
    count=pd.NamedAgg(column="Attrition", aggfunc="count")
).reset_index()


groupby_attrition_gender["percent"] = groupby_attrition_gender["count"] /  groupby_attrition_gender.groupby(["Gender"])["count"].transform('sum') *100

groupby_attrition_gender = groupby_attrition_gender.loc[groupby_attrition_gender['Attrition'] == "Yes"]

groupby_attrition_gender.head()
```

```python
fig = px.bar(groupby_attrition_gender, x="Attrition", y = "percent", color="Gender", width=800, height=500, barmode="group")
fig
```

*   This visualisation is comparing the attrition rate (in percentages) between male and female using barchart. Male is coloured in red and female is coloured in blue.
*   There is only a marginal difference in attrition rate between male and female. The attrition is about 17% for male and 14.8% for female.In summary, the attrition rate for male is slightly higher compared to female.


1.3 What is the distribution of daily rates across different departments?

```python
groupby_dailyrate_dpt = df_attrition[["DailyRate", "Department"]].groupby(["Department"], as_index=False).mean()
groupby_dailyrate_dpt
```

```python
fig = px.bar(groupby_dailyrate_dpt, x="Department", y="DailyRate", width=800, height=500, barmode="group")
fig.update_layout(
    title="Comparing average daily rate between departments",
    xaxis_title="Department",
    yaxis_title="Average daily rate",
)
fig
```

1.   This visualisation is comparing the attrition rate (in percentages) between departments using barchart.
2.   There is only marginal differences between the 3 departments. Research and development gets paid the most, Sales comes second then Human Resources at last.



```python
fig = px.box(df_attrition, x="Department", y="DailyRate", color="Department")
fig.update_layout(
    title="Comparing daily rate between departments",
    xaxis_title="Department",
    yaxis_title="Daily rate",
)
fig
```

1.   This visualisation is comparing the distrubition of daily rate (in percentages) between departments using boxplot. Colouring has been used to discriminate between departments. 
2.   According to the box plot, the three departments have very similary minimum, maximum and median for daily rate.



```python
fig = px.histogram(df_attrition, x="DailyRate", facet_row="Department", color="Department",
                   opacity=0.5, nbins=28, histnorm='percent',
                   title="Distribution of daily rate between departments")
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[0]))
fig
```

1.   This visualisation is comparing the distrubition of daily rate (in percentages) between departments using histogram. Colouring has been used to discriminate between departments. The graph is faceted by department.
2.   According to the histogram, the 3 departments have very similar distrubitions, there are only a few negligible differences. 1 thing to notice is that the number of HR employees is a lot smaller than the other 2 departments, therefore the variability of daily rate is bigger.


1.4 What factors lead to employee attritions? Focusing on four columns: Department, OverTime, JobLevel, and JobSatisfaction

```python
groupby_attr_dpt = df_attrition.groupby(["Attrition", "Department"]).agg(
    count=pd.NamedAgg(column="Attrition", aggfunc="count")
).reset_index()

groupby_attr_dpt["percent"] = groupby_attr_dpt["count"] /  groupby_attr_dpt.groupby(["Department"])["count"].transform('sum') *100

groupby_attr_dpt = groupby_attr_dpt.loc[groupby_attr_dpt['Attrition'] == "Yes"]

groupby_attr_dpt
```

```python
fig = px.bar(groupby_attr_dpt, x="Attrition", y="percent", color="Department", barmode = "group")
fig
```

1.   This visualisation is comparing the attrition rate (in percentages) between departments using barchart. Colouring has been used to discriminate between departments.
2.   The rate of attrition accross three departments are generally below 21%. However, research and developement has the lowest attrition rate at about 14% as supposed to sales with highest attrition rate at 20.6% and HR at about 19%. 



```python
groupby_attr_ot = df_attrition.groupby(["Attrition", "OverTime"]).agg(
    count=pd.NamedAgg(column="Attrition", aggfunc="count")
).reset_index()

groupby_attr_ot["percent"] = groupby_attr_ot["count"] /  groupby_attr_ot.groupby(["OverTime"])["count"].transform('sum') *100

groupby_attr_ot = groupby_attr_ot.loc[groupby_attr_ot['Attrition'] == "Yes"]

groupby_attr_ot
```

```python
fig = px.bar(groupby_attr_ot, x="Attrition", y="percent", color="OverTime", barmode = "group")
fig
```

1.   This visualisation is comparing the attrition rate (in percentages) between overtime categories using barchart. Colouring has been used to discriminate between overtime categories.
2.   It is clear that attrition rate is about 3 times higher for those who have to work overtime.



```python
df_attrition['JobLevel']=df_attrition.JobLevel.astype('category')
```

```python
groupby_attr_jl = df_attrition.groupby(["Attrition", "JobLevel"]).agg(
    count=pd.NamedAgg(column="Attrition", aggfunc="count")
).reset_index()

groupby_attr_jl["percent"] = groupby_attr_jl["count"] /  groupby_attr_jl.groupby(["JobLevel"])["count"].transform('sum') *100

groupby_attr_jl = groupby_attr_jl.loc[groupby_attr_jl['Attrition'] == "Yes"]

groupby_attr_jl
```

```python
fig = px.bar(groupby_attr_jl, x="Attrition", y="percent", color="JobLevel", barmode = "group")
fig
```

1.   This visualisation is comparing the attrition rate (in percentages) between job level categories using barchart. Colouring has been used to discriminate between job level categories.
2.   Generally speaking, the lower the job level, the higher the attrition rate. Job level 1 has the highest attrition rate at about 26%, followed by job level 3, level 2 and level 5 in descending order. Job level 4 has the lowest attrition rate.



```python
df_attrition['JobSatisfaction']=df_attrition.JobSatisfaction.astype('category')
```

```python
groupby_attr_js = df_attrition.groupby(["Attrition", "JobSatisfaction"]).agg(
    count=pd.NamedAgg(column="Attrition", aggfunc="count")
).reset_index()

groupby_attr_js["percent"] = groupby_attr_js["count"] /  groupby_attr_js.groupby(["JobSatisfaction"])["count"].transform('sum') *100

groupby_attr_js = groupby_attr_js.loc[groupby_attr_js['Attrition'] == "Yes"]


groupby_attr_js
```

```python
fig = px.bar(groupby_attr_js, x="Attrition", y="percent", color="JobSatisfaction", barmode = "group")
fig
```

1.   This visualisation is comparing the attrition rate (in percentages) between job satisfaction categories using barchart. Colouring has been used to discriminate between job satisfaction categories.
2.   Generally speaking, the higher the job satisfaction, the lower attrition rate is. Those with low job satisfaction has the highest attrition rate at about 23%. Those with medium and high job satisfaction rate are at a tie at about 16.5% attrition rate. Those with very high job satisfaction has the lowest attrition rate at only 11.3%.
