### **Dataset description** 

Soccer match event dataset provides to the public the largest open collection of soccer-logs ever released, collected by Wyscout (https://wyscout.com/) containing all the spatio-temporal events (passes, shots, fouls, etc.) that occur during all matches of an entire season of seven competitions (La Liga, Serie A, Bundesliga, Premier League, Ligue 1, FIFA World Cup 2018, UEFA Euro Cup 2016). A match event contains information about its position, time, outcome, player and characteristics.



### **Data collection**
Run the code below to load the datasets to this notebook. 

```python
!wget https://ndownloader.figshare.com/files/14464685/events.zip
import pandas as pd
import zipfile

df_players = pd.read_json("https://ndownloader.figshare.com/files/15073721")

with zipfile.ZipFile("events.zip") as z:
   with z.open("events_Spain.json") as f:
      df_events = pd.read_json(f)
```

The code cell above will download two datasets from the archive:


*   Player dataset (`df_players`): describes all players of the teams playing in seven national and international soccer competitions (Italian, Spanish, French, German, English first divisions, World Cup 2018, European Cup 2016). See this [link](https://figshare.com/articles/dataset/Players/7765196?backTo=/collections/Soccer_match_event_dataset/4415000) for details description of each attribute.
*   Even dataset (`df_events`): describes all the events that occur during each match. Each event refers to a ball touch and contains the information such as players involved, event names, and event times. See this [link](https://figshare.com/articles/dataset/Events/7770599?backTo=/collections/Soccer_match_event_dataset/4415000) for details description of each attribute.



```python
# extract data for nationality, full name, and their position in the team
df_players["Nationality"] = [val["name"] for val in df_players["passportArea"]]
df_players["Full Name"] = df_players["firstName"] + " " + df_players["lastName"]
df_players["position"] = [val["name"] for val in df_players["role"]]

df_players.head()
```

```python
# show the first 5 rows of the event table
df_events.head()
```

Below in the basic code cell to merge the two datasets for analyses. Columns `playerId` in the `df_events` dataframe and `wyId` in the `df_players` dataframe both are the unique IDs of players in the datasets. The created dataframe `df_all` will be the data used in your tasks.

```python
df_all = df_events.merge(df_players[["wyId", "Full Name", "Nationality", "height", "weight", "position"]], left_on="playerId", right_on="wyId")
df_all.head()
```

2.1 Create visualisation(s) to rank players based on the number of shots they made. Highlight the top 5 players with the most shots. (5 marks)

```python
df_all_shot = df_all.loc[df_all['eventName'] == "Shot"]

df_all_shot_groupby = df_all_shot.groupby(["playerId","Full Name","weight","height","Nationality"]).agg(
    count=pd.NamedAgg(column="eventName", aggfunc="count")
).reset_index()

df_all_shot_groupby["playerId_cat"] = df_all_shot_groupby["playerId"].astype("string")

df_all_shot_groupby=df_all_shot_groupby.sort_values('count', ascending = False)

df_all_shot_groupby
```

```python
df_top10 =  df_all_shot_groupby.head(10)
```

```python
df_top5 =  df_all_shot_groupby.head(5)
df_top5
```

```python
top5list=df_top5['Full Name'].tolist()
print(top5list)
```

```python
fig = px.bar(df_top10, x="Full Name", y="count",
                labels={
                     "playerId_cat": "Player ID",
                     "count": "Shot"
                     
                 }, hover_name= "Full Name", hover_data=["weight","height","Nationality"])
fig.update_layout(
    title={
        'text': "Ranking of players based on number of shots",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig["data"][0]["marker"]["color"] = ["red" if playerId_cat in top5list else "blue" for playerId_cat in fig["data"][0]["x"]]
fig
```

1. This visualisation is ranking the players based on number of shots using barchart. Red colouring has been used to highlight the top 5 players with most shots out of the top 10 players. Hovering over the individual bars will reveal the player's details such as height, weight, nationality and shot.

compares the number of

```python
df_all_position = df_all.groupby(["position"]).agg(
    count=pd.NamedAgg(column="position", aggfunc="count")
).reset_index()
df_all_position
```

```python
df_all_pass = df_all.loc[df_all['eventName'] == "Pass"]

df_all_pass_position_groupby = df_all_pass.groupby(["position","eventName"]).agg(
    count=pd.NamedAgg(column="position", aggfunc="count")
).reset_index()

df_all_pass_position_groupby["Percentage"] = round((df_all_pass_position_groupby["count"] / df_all_position.groupby(["position"])["count"].transform('sum') *100),2)

df_all_pass_position_groupby
```

```python
fig_pass_position = px.bar(df_all_pass_position_groupby, x="position", y = "Percentage", color="position")
fig_pass_position.update_layout(
    title={
        'text': "Compare passes across all positions",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig_pass_position
```

1.   This visualisation is comparing passes (in percentages) using barchart. Colouring has been used to discriminate between the all the positions.
2.   According to the graph, on average, a defender makes the most passes at 58.02%, followed by a midfiedler at 57.83%. A goal keeper comes third at 48.72% and followed by a forward at 41.74%.

```python
df_all_shot = df_all.loc[df_all['eventName'] == "Shot"]

df_all_shot_position_groupby = df_all_shot.groupby(["position","eventName"]).agg(
    count=pd.NamedAgg(column="position", aggfunc="count")
).reset_index()

df_all_shot_position_groupby["Percentage"] = round((df_all_shot_position_groupby["count"] / df_all_position.groupby(["position"])["count"].transform('sum') *100),2)

df_all_shot_position_groupby
```

```python
fig_shot_position = px.bar(df_all_shot_position_groupby, x="position", y = "Percentage", color="position")
fig_shot_position.update_layout(
    title={
        'text': "Compare shots across all positions",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig_shot_position
```

1.   This visualisation is comparing shots (in percentages) using barchart. Colouring has been used to discriminate between the all the positions.
2.   According to the graph, on average, a forward makes the most shots at 3.57%, followed by a midfiedler at 1.41%. A defender comes third at 0.56% and followed by a goal keeper at 0.01%.
